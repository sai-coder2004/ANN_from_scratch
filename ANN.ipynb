Here we implement an Artificial Neural network form scratch without any predefined libraries

Regression Problem

```python
import numpy as np
import pandas as pd
import random

df = pd.DataFrame([[8,8,4], [7,9,5], [6,10,6], [5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])

df
```

```python
def initialize_params(weights):
    wei1=np.full((weights[0],weights[1]),0.01)
    bias1=np.zeros(weights[1])
    wei2=np.full((weights[1],weights[2]),0.01)
    bias2=np.zeros(weights[2])
    return wei1,bias1,wei2,bias2
```

```python
w1,b1,w2,b2=initialize_params([2,2,1])
```

```python
b2[0]
```

```python
def fwd_propagation(X,w,b):
    layer1_op=np.dot(w.T,X)+b
    return layer1_op
```

```python
def MSE(y,y_hat):
    return (y-y_hat[0][0])**2
```

```python
df1=df.iloc[:,0:2]
Y=df.iloc[:,-1]
df1 = (df1 - df1.mean()) / df1.std()
df1
```

```python
df2=np.array(df1)
df2
```

```python
def update_weights(y,y_hat,j,op1):
    #print(b2[0])
    b2[0]=b2[0]+0.01*2*(y-y_hat[0][0])
    #print(y_hat[0][0])
    w2[0][0]=w2[0][0]+1e-4*2*(y-y_hat[0][0])*op1[0][0]
    w2[1][0]=w2[1][0]+1e-4*2*(y-y_hat[0][0])*op1[1][0]
    w1[0][0]=w1[0][0]+1e-4*2*(y-y_hat[0][0])*w2[0][0]*df2[j][0]#w11
    w1[0][1]=w1[0][1]+1e-4*2*(y-y_hat[0][0])*w2[1][0]*df2[j][0]#w12
    w1[1][0]=w1[1][0]+1e-4*2*(y-y_hat[0][0])*w2[0][0]*df2[j][1]#w21
    w1[1][1]=w1[1][1]+1e-4*2*(y-y_hat[0][0])*w2[1][0]*df2[j][1]#w22
    b1[0]=b1[0]+1e-4*2*(y-y_hat[0][0])*w2[0][0]#b11
    b1[1]=b1[1]+1e-4*2*(y-y_hat[0][0])*w2[1][0]#b22
    
    
```

```python
for i in range(0,10000):
    loss=0;
    for j in range(0,df2.shape[0]):
        idx=random.randint(0,df2.shape[0]-1)
        op1=fwd_propagation(df2[idx].reshape(2,1),w1,b1.reshape(2,1))
        #print(op1)
        y_hat=fwd_propagation(op1,w2,b2.reshape(1,1))
        #print(y_hat)
        loss=loss+MSE(Y[idx],y_hat)
        update_weights(Y[idx],y_hat,idx,op1)
    print("epoch: ",i,loss//df2.shape[0])
        
```
